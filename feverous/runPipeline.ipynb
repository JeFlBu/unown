```python
import os
import cell_retriever
import predictor_universal
import json
import matplotlib.pyplot as plt
import numpy as np
import math
import time
import torch
import shutil
import traceback

#Root directory, where your files are located
base_path='/homes//feverous_work/feverousdata' #Your working directory


#Choose the correct cell depending on your setting (table+text or text only)
# And put in names_to_use the names of your experiments directories
```

```python
def run_exp_multitest(exp_folder , dest_folder, train_db, train_dev_file, dev_db, dev_testfile, suffix ,pretrained_retriever_to_use,pretrained_predictor_to_use,usinggt, usingretriever):

    list_exp_names = os.listdir(base_path+'/'+exp_folder+'/')
    list_exp_names2 = [x for x in list_exp_names if '.jsonl' in x]
    list_files_to_add = [base_path+'/' +exp_folder+'/'+ x for x in list_exp_names2]
    list_exp_names=[]
    
    for i in range(len(dest_folder)):
        list_exp_names += [[dest_folder[i]+'/' + x.split('.')[0] for x in list_exp_names2]]
        if not os.path.exists(base_path+'/'+dest_folder[i]):
            os.mkdir(base_path+'/'+dest_folder[i])
            
    for ft in enumerate(list_files_to_add):
        
        db_path=base_path+'/'+train_db


        train_devset=train_dev_file

        data_path=base_path
        f2=open(ft[1],'r')
        f3=open(base_path+"/train.jsonl",'w')
        print(ft[1])
        for line in f2:

            f3.write(line)
        f3.close()
        f_params=open(ft[1].split('.json')[0]+'_params.json','r')
        obj_params=json.load(f_params)
        f_params.close()

        

        for i in range(len(dest_folder)):
            exp_name=list_exp_names[i][ft[0]]
            dev_db_path=base_path+'/'+dev_db[i]
            test_devset=dev_testfile[i]
            
            if os.path.exists(data_path+"/"+exp_name+'_predictor_scores.json') :
                print("|i> Predictor "+ft[1]+" for "+ dest_folder[i]+" already done .")
                continue
            start = time.time()
            predictor_universal.exec_predictor_alt_devdb_quick(db_path, data_path, exp_name,dev_db_path,train_devset, test_devset, False, usinggt=usinggt, usingretriever=usingretriever)
            


            end = time.time()
            obj_params['execution_time']=end - start
            f_params=open(data_path+"/"+exp_name+'_predictor_params.json','w')
            json.dump(obj_params,f_params)
            f_params.close()
        else:
            print("|i> "+ft[1]+" already done .")
           
```

```python
#sentences only setting


names_to_use=[]#Put here the list of directories containing experiments (*.jsonl+*_params.json)

for name_to_use in names_to_use:
    run_exp_multitest(name_to_use, ['res_'+name_to_use+'_s_original'], 'jf_home/feverous_wikiv1.db', 'feverous_dev_challenges_sentencesonly.jsonl',['jf_home/feverous_wikiv1.db'], ['feverous_dev_challenges_sentencesonly.jsonl'],'', None, None, True, True)
    
```
